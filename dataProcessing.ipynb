# 6. What are the 10 neighborhoods with the highest average sale price per square meter? What are the 10 neighborhoods with the highest average rental price?

```python
#Import the data
df = pd.read_excel('realEstateDataAssessment.xlsx')
```

```python
#Nomalize the neighborhood data
df['barrio'] = df['barrio'].str.lower()
df.head()
```

```python
#Get neigborhoods that have regions as a value
df_nei = df.loc[df['barrio'].isin(['nororiente', 'suroriente', 'occidente','suroccidente',
       'noroccidente', 'centro'])]
df_nei.head()
```

```python
#Get rows number of neigborhoods that have regions as a value cases
rows_index = []
for row in df_nei.index:
    rows_index.append(row)
print(rows_index)
```

```python
#Drop neigborhoods that have regions as a value cases
for i in rows_index:
    df = df.drop(i)
df.head()
```

```python
#Get the unique neigborhood names
nei = df['barrio'].unique()
status = df['oferta'].unique()

top_dict_rent = {}
top_dict_sale = {}

for i in nei:
    for j in status:
        temp_nei = 0
        temp_mean = 0
        if j == 'Arriendo':
            ans = df.query("barrio == @i and oferta == @j and precio < 18e6")
            mean_t = ans['precio'].mean()
            temp_nei = i
            temp_mean = mean_t
            top_dict_rent[temp_nei] = temp_mean
    else:
            ans = df.query("barrio == @i and oferta == @j")
            mean_t = ans['precio'].mean()
            temp_nei = i
            temp_mean = mean_t
            top_dict_sale[temp_nei] = temp_mean
```

```python
#Get the top 10 most expensive rents by neighborhood
top_10_rent = nlargest(10, top_dict_rent, key = top_dict_rent.get)
cnt = 1
for i in top_10_rent:
    print("El top " + str(cnt) +" mas costoso en renta son " + i + " --> " + str(top_dict_rent[i]))
    cnt += 1
```

```python
#Get the top 10 most expensive sales by neighborhood
top_10_sale = nlargest(10, top_dict_sale, key = top_dict_sale.get)
cnt = 1
for i in top_10_sale:
    print("El top " + str(cnt) + " mas costoso en venta son " + i + " --> " + str(top_dict_sale[i]))
    cnt += 1
```
